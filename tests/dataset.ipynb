```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import config
```

```python
data = pd.read_csv(config.DATASET_PATH)
```

```python
data.select_dtypes(exclude=object).hist()
plt.show()
```

```python
numerical = data.select_dtypes(exclude=object)
normalized = (numerical - numerical.mean(axis=0)) / numerical.std(axis=0)
for column in normalized.columns:
    plt.figure()
    normalized[column].hist()
    plt.show()
```

```python
for column in data.select_dtypes(include=object).columns:
    plt.figure()
    plt.title(f'Value counts for {column} features')
    data[column].value_counts().plot(kind='bar')
    plt.show()
```
